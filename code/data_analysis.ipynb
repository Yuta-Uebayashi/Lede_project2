## データを統合

```python
df = pd.read_csv("2020_census_cleaned.csv")
df.head(15)
```

```python
#２０２０_censusに2010_censusの「外国人比率」のデータを統合したい。自治体のコードが一致するものを統合する。
# 2010年のデータを読み込む
df_2010 = pd.read_csv("2010_census_cleaned.csv")

# 2010年の「外国人比率」列だけ抽出（カラム名は適宜変更）
df_2010_ratio = df_2010[['自治体コード', '外国人比率（％）']].copy()
df_2010_ratio = df_2010_ratio.rename(columns={'外国人比率（％）': '2010年外国人比率（％）'})

# 2020年データとマージ（自治体コードで結合）
df = df.merge(df_2010_ratio, on='自治体コード', how='left')
```

```python
df.head(15)
```

```python
df = df.set_index('全国，都道府県，市区町村')
df.head(15)
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head(15)
```

```python
df.to_csv("census_merged.csv")
```

```python
#自治体コードを５桁に変換
import pandas as pd

# CSVファイルを読み込み
df = pd.read_csv("census_merged.csv")

# 自治体コードをゼロ埋めして文字列に変換（5桁）
df["自治体コード_文字列"] = df["自治体コード"].astype(str).str.zfill(5)

# 変換結果を確認（上から5行だけ表示）
print(df[["自治体コード", "自治体コード_文字列"]].head())

# 必要なら新しいCSVとして保存
df.to_csv("census_merged_fixed.csv", index=False)
```

# Japanese population

```python
import pandas as pd
```

```python
# データの読み込み
df = pd.read_csv('only_Japanese_population.csv')
df.head(15)
```

```python
df.shape
```

```python
df.info()
```

```python
df['population'] = df['population'].str.replace(',', '')
df['population'] = pd.to_numeric(df['population'], errors='coerce')
```

```python
import matplotlib.pyplot as plt

ax = df.plot(kind='bar', x='year', y='population', figsize=(18, 8))
ax.set_title('Japanese Population')      # グラフのタイトル
ax.set_ylabel('people')            # Y軸ラベル
plt.show()
```
